```python
flightdetails['FL_PUSH_DATE'] = pd.to_datetime(flightdetails['FL_PUSH_DATE'])
flightdetails['FL_SCHEDULE_DATE'] = pd.to_datetime(flightdetails['FL_SCHEDULE_DATE'])
flightdetails['FL_MESSAGE_DATE'] = pd.to_datetime(flightdetails['FL_MESSAGE_DATE'])
```

```python
flightdetails['FL_SCHEDULE_YEAR'] = flightdetails['FL_SCHEDULE_DATE'].dt.year
flightdetails['FL_SCHEDULE_MONTH'] = flightdetails['FL_SCHEDULE_DATE'].dt.month
flightdetails['FL_SCHEDULE_DAY'] = flightdetails['FL_SCHEDULE_DATE'].dt.day
flightdetails['FL_SCHEDULE_DAYOFWEEK'] = flightdetails['FL_SCHEDULE_DATE'].dt.dayofweek
flightdetails['FL_SCHEDULE_WEEKDAY'] = flightdetails['FL_SCHEDULE_DATE'].dt.weekday
```

```python
flightdetails[flightdetails.FL_SCHEDULE_YEAR==2017].groupby(['FL_SCHEDULE_YEAR','FL_SCHEDULE_MONTH'])['FL_AFSKEY'].count()\
                                                                    .reset_index(name="COUNT").sort_values(by=['COUNT'],ascending=False)
```

```python
flightdetails[(flightdetails.FL_SCHEDULE_YEAR==2017) & (flightdetails.FL_SCHEDULE_MONTH==12) ].groupby(['FL_SCHEDULE_YEAR','FL_SCHEDULE_MONTH','FL_SCHEDULE_DAY','FL_SCHEDULE_WEEKDAY'])['FL_AFSKEY'].count()\
                                                                    .reset_index(name="COUNT").sort_values(by=['COUNT'],ascending=False)
```
